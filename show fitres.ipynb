```python
res = np.load('fitres1.npy')
names = ['WV', 'alpha', 'beta', 'g_dsa', 'g_dsr']
lims = [(0,.5), (0,3), (0,.5), (0,80), (0,80)]
```

```python
figsize(15,8)
for i, d in enumerate(zip(names, np.rollaxis(res,2), lims)):
    n,v,l = d
    subplot(3,2,i+1)
    title(n)
    imshow(np.rot90(v,2))
    if l is not None:
        clim(*l)
    colorbar()
tight_layout()
```

```python
from runmacs.spec.util.specrend import SpecRGBConverter
from runmacs.spec.util.sensorparameters import sensorParameters
from downwelling import DownwellingModel
from atmospheredata import AtmosphereData
wvlns = np.array(sensorParameters['vnir']['loadWavelength']())[::2]
```

```python
m = DownwellingModel(AtmosphereData(wvlns=wvlns))

m.setConstant('f_dd', 0)
m.setConstant('f_ds', 1)
m.setConstant('H_oz', 0.4)
#m.setConstant('alpha', 1.4)
m.setConstant('wvl_a', 550)
m.setConstant('RH', .7)
m.setConstant('P', 1013.25)
m.setConstant('AM', 3)

#m.setConstant('theta_sun', np.deg2rad(46.068404))
m.setConstant('theta_sun', np.deg2rad(40))
```

```python
names, function = m.getLambdifiedE()
```

```python
spectralImage = function(wvlns, *[res[...,i,np.newaxis] for i in range(len(names))]) 
```

```python
rend = SpecRGBConverter(wvlns)
```

```python
rgbImage = rend.spectrum_to_rgb(spectralImage)
```
